```python
import pandas as pd
import seaborn as sns
import numpy as np
import pylab as plt
```

```python
df = pd.read_excel(r"C:\\Users\\aakan\\OneDrive\\Documents\\BE\\BDA\\Miniproj\\Data_Train.xlsx")
```

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])
df['Date_of_Journey'] = df['Date_of_Journey'].dt.month
```

```python
#No of flights in each month
months = df.Date_of_Journey.value_counts()
sns.barplot(x=months.index, y=months.values)
plt.show()
```

```python
#No of flights of each airline
airline = df.Airline.value_counts()
#print(airline)
plt.figure(figsize=(18,4))
sns.barplot(x=airline.index, y=airline.values)
plt.xticks(rotation=45, fontsize=8)
plt.show()
```

```python
source = df.Source.value_counts()
dest = df.Destination.value_counts()

plt.subplot(121)
sns.barplot(x=source.index, y=source.values)
plt.title('Most frequent origins')
plt.xlabel('Source')
plt.ylabel('Number of take-offs')
plt.xticks(rotation=45)

plt.subplot(122)
sns.barplot(x=dest.index, y=dest.values)
plt.title('Most frequent destinations')
plt.xlabel('Destination')
plt.ylabel('Number of landings')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#No of flights per origin
n_flights_per_origin = df.Source.value_counts()
n_flights_per_origin
print('Total number of flights : ', n_flights_per_origin.sum())
print('Percentage contribution of flights from Delhi : ', (n_flights_per_origin['Delhi']/n_flights_per_origin.sum())*100)
```
